```python
#As there are more outliers I have chosen median over mean to fill the null data.
Olympics['Age']=Olympics['Age'].fillna(Olympics['Age'].median())
Olympics['Height']=Olympics['Height'].fillna(Olympics['Height'].median())
Olympics['Weight']=Olympics['Weight'].fillna(Olympics['Weight'].median())
```

```python
HOW MANY GOLD,SILVER,BRONZE MEDALS HAVE BEEN AWARDED TO MALE AND FEMALE ATHELETES?
```

```python
con=pd.crosstab(Olympics['Sex'],Olympics['Medal'])
con
```

```python
sns.heatmap(con,annot=True,fmt=".0f",cmap='coolwarm')
plt.xlabel("Medal",fontsize=12,fontweight='bold')
plt.ylabel("Gender",fontsize=12,fontweight='bold')
plt.title("Gender-Medal Distribution in Olympics",fontsize=14,fontweight='bold')
```

```python
DISPLAY MEDAL PERFORMANCE FOR A SPECIFIC COUNTRY PER YEAR
```

```python
df_India=Olympics[Olympics['Team']=='India']
```

```python
Medal_India=df_India.groupby(['Year'])['Medal'].count()
Medal_India.values
```

```python
sns.lineplot(x=Medal_India.index,y=Medal_India.values)
plt.xlabel("Year",fontsize=12,fontweight='bold')
plt.ylabel("Count",fontsize=12,fontweight='bold')
plt.title("INDIA-Medal Distribution in Olympics",fontsize=14,fontweight='bold')
```

```python
FIND AND DISPLAY THE COUNT OF MEDALS WON IN EACH CITY AND PRESENT THE RESULT IN TABLE FORMAT
```

```python
table=pd.crosstab(Olympics['City'],Olympics['Medal'])
table
```
